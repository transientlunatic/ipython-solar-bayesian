For the next stage we need to define a threshold (we have, in fact, done this already, although implicitly for graphing the highlights). The `thresholder` method clips the odds ratio array at a given value (here 7.027591): any value below this is not a flare, and any region with values above it is a flare candidate, and proceeds to the next step of the analysis. The `expand` argument pads the candidates, so very close regions will be combined to one.

The value used for the thresholding is calculated by iterating the analysis over a very large number (in this case 2000) Lightcurve objects containing simulated data---data which is sampled from a Gaussian distribution, and so represents a simulation of Gaussian noise. The maximum log(odds ratio) value is then found for each of these 2000 lightcurves, and these values are then ordered, with the thresholding value being the one representing the 99th percentile, in the case of a desire to detect with a false-alarm probability of 1 percent.

```python
flarelist, Nflares, maxlist = odds.thresholder(odds_array, 7.027591, expand=3)
```

The `flare_list_dataframe` function is defined just for this notebook. In the full analysis results are stored into a database, and the `bayesflare` database interface performs this function's job. 

The current implementation of the database used in Bayesflare relies on SQLite3, without using any form of ORM mapping. This is probably *not* a good long-term solution, and a better solution would be to port this class to an SQLAlchemy declarative approach.

The flarelist generated by the thresholder above is input to the function, which generates a dataframe containing details of all of the flares.

```python
def flare_list_dataframe(flarelist, maxlist):
    my_flares = []
    i = 0
    for flare in flarelist: 
        my_flares.append({
            'start_time':data.data.index[flare[0]],
            'peak_time':data.data.index[maxlist[i][1]],
            'end_time':data.data.index[flare[1]],
            'class':data.goes_class(data.data.xrsb[maxlist[i][1]])
        })
        i+=1
    my_flares = pd.DataFrame(my_flares)
    return my_flares

def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return idx, array[idx]
```

```python
def flare_boundaries(t0, t_gauss, t_exp, amount = 0.95):
    """
    Calculates the start and end times of a flare given a midpoint, 
    a t_gauss value, and a t_exp value.
    
    Parameters
    ----------
    t0 : datetime.datetime object
       The midpoint time of the flare.
    t_gauss : float
       The gaussian rise parameter. In seconds.
    t_exp : float
        The exponential decay factor. In seconds.
    amount : float
        The amount of the flare which the start and 
        end time should enclose. Defaults to 0.95 (per cent).
    
    Returns
    -------
    start : datetime.datetime object
        The datetime of the start of the flare.
    end : datetime.datetime object
        The datetime of the end of the flare.
        
    """

    ramount = 1 - amount
    
    gauss_cut = sqrt(abs(log(ramount) * 2 * gauss**2))
    exp_cut = abs(exp*log(ramount)) 
    start = t0 - datetime.timedelta(seconds=gauss_cut)
    end = t0 + datetime.timedelta(seconds=exp_cut)

    return start, end
```
