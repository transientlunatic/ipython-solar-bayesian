Infuriatingly, the SunPy interface only downloads the data, and doesn't perform the corrections, so we need to pull the data out of IDL.

```python
corr_data = fits.open('data/hsi_crtd_20100804_000000_48hrs.fits')
```

```python
data = corr_data[1].data[0][0]
time = [datetime.datetime.strptime(a, '%Y-%m-%dT%H:%M:%S') for a in data]
#pl.plot(x, corr_data[1].data[0][1])
#data = np.transpose(corr_data[1].data[0][1])
data = corr_data[1].data[0][1]
data = pd.DataFrame(data, index=time, columns=[u'3 - 6 keV', u'6 - 12 keV', u'12 - 25 keV', u'25 - 50 keV', u'50 - 100 keV', u'100 - 300 keV', u'300 - 800 keV', u'800 - 7000 keV', u'7000 - 20000 keV'])
```

```python
# Behold, a most dirty of hacks
start = '2014-08-04 00:00'
end = '2014-08-06 00:00'
rhessi = lc.RHESSISummaryLightCurve.create(start, end)
rhessi.data = data
```

```python
fig = plt.figure()
rhessi.data['6 - 12 keV'].plot(fig = fig, figsize=(18, 7))
```

```python
import pylightcurve as lc
```

```python
curve = lc.Lightcurve()
```

```python
meta = {}
for column in rhessi.data.columns:
    meta[column] = rhessi.meta
curve.import_data(rhessi.data, meta)
curve.clc = curve.data['6 - 12 keV']
```

```python
dcurve = curve.detrend(method='savitzkygolay', nbins=3001, order=3, column='6 - 12 keV')
dcurve = dcurve.detrend(method='savitzkygolay', nbins=3001, order=3, column='3 - 6 keV')
dcurve.data = dcurve.data.drop([u'12 - 25 keV', u'25 - 50 keV', u'50 - 100 keV', u'100 - 300 keV', u'300 - 800 keV', u'800 - 7000 keV', u'7000 - 20000 keV'], axis=1)
```

```python
dcurve.peek()
```

```python
import copy
import pandas as pd
def decimate(lightcurve):
    dl = copy.deepcopy(lightcurve)
    dl.data = pd.DataFrame()
    dl.data = lightcurve.data[::10]
    dl.cts = dl.cts[::10]
    
    return dl
```

```python
dcurve_d = decimate(dcurve)
curve_d = decimate(curve)
```

```python
len(curve_d.cts)
```

```python
bglen = 301
noiseestmethod='powerspectrum'
psestfrac=0.5
flareparams={'taugauss': (0,  10, 5), 
             'tauexp'  : (0, 150, 60)}
odds = bf.OddsRatioDetector( 
                            curve_d,
                            bglen=bglen,
                            bgorder=3,
                            noiseestmethod='powerspectrum',
                            tvsigma=1,
                            psestfrac=0.5,
                            flareparams=flareparams,
                            noisepoly=True,
                            noiseimpulse=True,
                            noiseimpulseparams={'t0': (0, (bglen-1.)*curve.dt(), bglen)},
                            noiseexpdecay=False,
                            noiseexpdecayparams={'tauexp': (600, 3*60*60, 6)},
                            noiseexpdecaywithreverse=True,
                            ignoreedges=False,
                            edgelength=10
                            )
odds_out = odds.oddsratio()
```

```python
# Load all of the odds ratio data into the Lightcurve
times = dcurve.data.index.tolist()
odds_data = pd.DataFrame(index=times)
odds_array = odds_out[0]
#odds_array = np.pad(odds_out[0],(int(21/4),int(21/4)), mode='minimum')
odds_data['xrsa'] = odds_array

data = dcurve.add_highlight(odds_data)

curve = dcurve
```

```python
%config InlineBackend.figure_format = 'png'
fig, ax = plt.subplots(1,1,figsize=(13,4), frameon=False, dpi=300)

ax.set_xlabel('Time')
ax.set_yscale('log')
ax.plot(data.data.index, data.data['6 - 12 keV'], alpha=0)
ax.set_ylim(1e-9, 1e-5)
pl.setp( ax.xaxis.get_majorticklabels(), rotation=90 )
ymin, ymax = ax.get_ybound()
xmin, xmax = ax.get_xbound()

stuff = np.array(data.highlight.xrsa)
threshold = 7.027591
stuff[stuff<threshold] = -100
stuff[stuff>=threshold] = -99
image = np.zeros((100,len(stuff)))
image[:,:] = stuff

ax1 = ax.twinx()
ax1.patch.set_visible(False)
ax1.set_yscale('log')
#ax1.set_ylim(1e-9, 1e-5)
#ax1.set_ylim(9e-7, 1e-6)
#ax1.set_ylabel('1.0--8.0 $\AA$ | Watts m$^-2$')
ax1.set_yticks(())
#ax1.set_ylabel('0.5--4.0 $\AA$ | Watts m$^-2$')
ax1.plot(data.data.index, data.data['3 - 6 keV'], alpha=1, color='#9d261d')

ax2 = ax.twinx()#.twiny()
if not len(data.highlight) == 0:
    ax2.imshow(image, origin='lower', cmap='RdPu', extent=[xmin, xmax, ymin, ymax], alpha=.1)
ax2.set_yscale('log')
ax2.set_yticks(())
ax2.patch.set_visible(False)
#ax2.yaxis.set_ticks_position((''))
#ax2.tick_params(axis='y', colors='blue')
ax2.plot(data.data.index, data.data['6 - 12 keV'], color='#049cdb', alpha=1)

spines_to_remove = ['top', 'right']
for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
almost_black = '#262626'
spines_to_keep = ['bottom', 'left']
for spine in spines_to_keep:
    ax.spines[spine].set_linewidth(0.5)
    ax.spines[spine].set_color(almost_black)    
#ax.grid(True)
ax.yaxis.grid(True, which="major", linestyle='-', color=almost_black, linewidth=0.15)

if data.tags:
    data.plot_tags(ax, 'peak_time', 'class')
```
