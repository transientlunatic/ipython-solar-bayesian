```python
data = Lightcurve()
meta = {}
for column in eve.data.columns:
    meta[column] = eve.meta
data.import_data(eve.data, meta)
```

```python
import pandas as pd
container = data.data['q0ESP']
data = Lightcurve()
data.data = pd.DataFrame(container)
```

```python
import bayesflare as bf
```

```python
data.time_seconds()
```

```python
curve_d = data
curve_d.cts, curve_d.clc = data.time_seconds(), data.data.q0ESP
curve_d.set_default('q0ESP')
bglen = 301
noiseestmethod='powerspectrum'
psestfrac=0.5
flareparams={'taugauss': (0,  10, 5), 
             'tauexp'  : (0, 150, 60)}
odds = bf.OddsRatioDetector( 
                            curve_d,
                            bglen=bglen,
                            bgorder=3,
                            noiseestmethod='powerspectrum',
                            tvsigma=1,
                            psestfrac=0.5,
                            flareparams=flareparams,
                            noisepoly=True,
                            noiseimpulse=True,
                            noiseimpulseparams={'t0': (0, (bglen-1.)*curve_d.dt(), bglen)},
                            noiseexpdecay=False,
                            noiseexpdecayparams={'tauexp': (600, 3*60*60, 6)},
                            noiseexpdecaywithreverse=True,
                            ignoreedges=False,
                            edgelength=10
                            )
odds_out = odds.oddsratio()
```
